```python
import pandas as pd
from os import walk
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
def load_dump_and_display_performance(dump_path):
    df = pd.read_csv(dump_path)
    print("Results for " + str(dump_path))
    if "twitter" in dump_path:
        #dict1 = {'racism':2,'sexism':1,'none':0}
        print("Precision: " + str(metrics.precision_score(df["y_true"], df["y_pred"], average=None, labels=[0, 2, 1])))
        print("Recall: " + str(metrics.recall_score(df["y_true"], df["y_pred"], average=None, labels=[0, 2, 1])))
        print("F1: " + str(metrics.f1_score(df["y_true"], df["y_pred"], average=None, labels=[0, 2, 1])))
    else:
        print("Precision: " + str(metrics.precision_score(df["y_true"], df["y_pred"])))
        print("Recall: " + str(metrics.recall_score(df["y_true"], df["y_pred"])))
        print("F1: " + str(metrics.f1_score(df["y_true"], df["y_pred"])))
        
    print(classification_report(df["y_true"], df["y_pred"]))
    
    print("")
```

```python
for (dirpath, dirnames, filenames) in walk("dumps"):
    for filename in filenames:
        load_dump_and_display_performance("dumps/" + filename)
    break
```
